```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def formula(g, K=255):
  g= g.astype(np.float64)
  min_g = np.min(g)

  g_m= g-min_g
  max_g_m = np.max(g_m)

  g_s= (K* g_m) / max_g_m

  return g_s
```

```python
ImgA=Image.open('/content/c(1).jpg').convert ('L')
ImgA=np.asarray(ImgA)

plt.imshow(ImgA, cmap='gray', vmin=0, vmax=255)
```

```python
plt.hist(ImgA.flatten(), bins=50, color='black');
```

```python
m=64 #debe ser un valor medio en el centro de los puntos que tienen valor en el histograma.
E=5 #Controla que tan pronunciada es la curva, se

g = 1 / (1 + (m / (ImgA + 0.00001))**E)
plt.imshow(g, cmap='gray')
plt.title(f'E: {E}, M: {m}')

gs = formula(g, 255)
#plt.imshow(gs, cmap='gray', vmin=0, vmax=255)
#Image.fromarray(np.array(gs, dtype=np.uint8)).save("IMAGEN.png")
```
